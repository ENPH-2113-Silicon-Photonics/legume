```python
from chickpea import sbl
import legume
import numpy as np
```

```python
shape_builder = sbl.hex_d6_poly(N_vertices=3, a=1)


lattice = legume.Lattice('hexagonal')
cryst = legume.PhotCryst(lattice)

eps=12

cryst.add_layer(d=0.6, eps_b=eps)

params = {"lengths": [0.4,0.3, 0.15],
          "angles": [np.radians(5),np.radians(20),np.radians(10)]}

shape_builder.place_shape(phc=cryst, eps=1, x=0, y=0, **params)

legume.viz.eps(cryst.layers[-1],Nx=200, Ny=200)
```
