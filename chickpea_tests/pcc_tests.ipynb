```python
from chickpea.pcc import PhotonicCrystalCavity
import legume
```

```python
#testing!! first test crystal types, radii, displacements
rad1 = .2
rad2 = 0.4
supc = [24, 16]
thickness = 0.6
eps = 12.0

pcc3 = PhotonicCrystalCavity(crystal='H', radius=rad2, supercell_size=supc, thickness=thickness, eps=eps, n=4)
print(pcc3.__repr__())

pcc4 = PhotonicCrystalCavity(crystal='L', radius=rad2, supercell_size=supc, thickness=thickness, eps=eps, n=4, m=5)
print(pcc4.__repr__())


pccs=[pcc3,pcc4]
```

```python
for pcc in pccs:
    cavity = pcc.cavity()

    legume.viz.eps_xy(cavity, cbar=True, Nx=200, Ny=200)
```
